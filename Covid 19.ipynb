```python
import requests
import pandas as pd 
import numpy 
```

```python
req = requests.get("https://api.covid19india.org/v4/data.json")
```

```python
data = req.json()
states = list(data)
data1 = []
for i in states:
    data1.append(data[i]['total'])
data1
```

```python
df = pd.DataFrame(data1)
df.drop(columns='other',inplace = True)
df
```
